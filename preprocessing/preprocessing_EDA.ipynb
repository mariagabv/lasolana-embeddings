## La Gaceta de La Solana

The Gaceta de La Solana is a bi-monthly printed newspaper with a tradition spanning over 50 years. Certain editions have surpassed 100 pages, covering a wide array of topics pertaining to the political, societal, and economic aspects of La Villa de La Solana, located in Ciudad Real.

For our purposes, 205 PDF editions from the Ayuntamiento de La Solana and [Biblioteca Virtual de Castilla-La Mancha webpages](https://ceclmdigital.uclm.es/results.vm?q=parent:0002509276&lang=es&view=global) were downloaded and stored in a folder and converted into TXT files with the open-source command-line utility `pdftotext`.

```bash
for pdf_file in '$input_directory'*.pdf; do
    base_name=$(basename '$pdf_file; .pdf)
    output_text_file='$output_directory${base_name}.txt'
    pdftotext -enc UTF-8 '$pdf_file' '$output_text_file'
```

Although this tool did a better job than others when converting PDFs to text than other python libraries, the original files present a challenge due to the inclusion of advertisements, tables, and typical newspaper formatting that is difficult to convert. This has resulted in some noise in our data. We will address this during preprocessing, but some inconsistencies could not be resolved automatically.

### Preprocessing

````python
with open("all_gacetas.txt", "r") as file:
    gacetas = file.read()
````

We remove hyphens and spaces that were necessary in the orignal text display (usually double column).

````python
matches = re.findall(r".*\w-\n", gacetas)
print(len(matches))
````

````python
replaced_hyphen = re.sub(r"(\w)-\n+", r"\1", gacetas).lower()
# remove sentences 'Biblioteca Virtual (...)
matches_publicado = re.sub(r"biblioteca virtual.*\n", "", replaced_hyphen)
# remove unnecessary spaces.
gacetas_stripped = re.sub(r"\n+|\s+|\t+", " ", matches_publicado).strip()
# remove all hyphen but those between characters, as in 'garcía-abadillo'
surname_hyphen = re.sub(r"(?<!\w)-(?!\w)", "", gacetas_stripped)
other_hyphen = re.sub(r"-\s|\s-", "", surname_hyphen)
# remove noisy sentences, often a bad conversion of ads
ad_remove = re.sub(r"[~•·].*?\.", "", other_hyphen)
````

````python
len(ad_remove)
````

Next, we tokenize the data into sentences, as sentences is what the librarty `fasttext` is expecting as an input for training word embeddings. We prepared our own list of symbols and replaced all numbers by the token `[NUM]`.


````python
sentences_gaceta = preprocess_sentences(ad_remove)
````

The next snippet prints random sentences to inspect our data and check that most problems are solved.

````python
num = random.randrange(0, 156508)
print(sentences_gaceta[num])
````

Finally, the data is saved into a text file, separated by a newline character.

````python
# file_path = "all_gacetas_preprocessed.txt"
# with open(file_path, "w", encoding='utf-8') as file:
# for sentence in sentences_gace:
# file.write(sentence + "\n")
````

## La Solana News Dataset

La Solana News Dataset consists of 1070 news articles in Spanish extracted from the 'News' section of the [La Solana City Council official website](https://www.lasolana.es/laciudad/noticias). The  Python libraries `bs4`, `rllib.parse`, and `newspaper` were used to collect and structure the data. Finally, manually annotate the data into seven clases. For training the word embeddings, we will use the columns 'Title' and 'Text'.

### Data Extraction and Preprocessing

````python
dataset = pandas.read_csv("lasolana_dataset.csv")
````

````python
titles = dataset["Title"]
text = dataset["Text"]
````

Next, we will create a function that iteratively joins each title to its article in our txt file, so they appear one after another.

````python
result_list = []

thread = threading.Thread(target=interleave_lists, args=(titles, text, result_list))
thread.start()
thread.join()
````

This return a list of strings. Because we want a unique string to perform sentece tokenization, the strings will be concatenated with a white space.

````python
concatenated_string = " ".join(result_list).lower()
````

The punctuation is removed and the numbers are replaced with the number token.

````python
sentences_articles = preprocess_sentences(concatenated_string)
````

````python
num = random.randrange(0, 1000)
print(sentences_articles[num])
````

Finally, we save the preprocessed document in a txt file.

## Julián Simón Blogspot

Julián Simón held the position of town mayor in La Solana from 1983 to 1987 and he was also a teacher. Starting from 2012 until 2019, he authored a digital blog called '[Joaquin Costa. La Solana. Legado Bustillo](https://joaquincostalasolanalegadobustillo.blogspot.com/)' with the objective of promoting the history and cultural heritage of La Solana. The entries of this blog are characterized by their linguistic precision and use of specialized in-domain terminology related to the town's traditions and culture. This makes the data high-quality for our training purposes.

### Preprocessing

The same proceadure as in the last two resources will be followed:

- Cleaning of specific markdown (blogspot markdown in this case)
- Setence Tokenization
- Number replacement by token [NUM]
- Saving the strings into a TXT file

````python
with open("blogsimón_raw", encoding="utf-8") as txt:
    data = txt.read()
````

````python
# Remove blogspot markdown
matches_publicado = re.sub(r"Publicado.*\n", "", data)
matches_comentario = re.sub(
    r"\d Añadir un comentario|\d Ver comentarios", "", matches_publicado
)
matches_day = re.sub(r"\d\n", "", matches_comentario)
matches_month = re.sub(
    r"MAY|APR|JUN|JUL|SEP|NOV|AUG|OCT|JAN|DEC|FEB|MAR\n", "", matches_day
)

# Remove unexpected 2 at the beginning of some sentences
dos_word = re.sub(
    r"(?<!\S)2", "", matches_month
)  # matches the character "2" only if it's at the beginning of a line or preceded by whitespace

# Remove extra whitespaces
text_stripped = re.sub(r"\n+|\s+", " ", dos_word).strip()
data_processed = text_stripped.lower()
````

````python
sentences_blogspot = preprocess_sentences(data_processed)
````

````python
num = random.randrange(0, 1000)
print(sentences_blogspot[num])
````
